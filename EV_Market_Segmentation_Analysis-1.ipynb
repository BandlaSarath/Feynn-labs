```python
ev_two_numeric = ev_two.select_dtypes(['float', 'int'])
ev_two_numeric.head()
```

```python
#importing pca algorithm
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=5)
ev_two_numeric_pca = pca.fit_transform(ev_two_numeric)
ev_two_numeric_pca[:5]
print(pca.components_)
```

```python
def pca_plot(pca, df, x_axis=0, y_axis=1, scale=1):
    plt.scatter(pca.components_[x_axis]*scale, pca.components_[y_axis]*scale, color='black')
    for x,y,name in zip(pca.components_[x_axis]*scale, pca.components_[y_axis]*scale, df.columns):
        plt.text(x, y+0.02, name, color='blue')
        plt.annotate('', (x,y), (0,0), arrowprops=dict(arrowstyle='->', color='red'))
    plt.xlabel(f'PCA{x_axis+1}')
    plt.ylabel(f'PCA{y_axis+1}')
    plt.show()

```

```python
pca_plot(pca,ev_two_numeric, 1, 2)
```

From the above graph, we can say that many of the users were positively respond for the features of maintence cost and extra features

**Encoding the Ordinal Features**

```python
# Define the ordinal features to be encoded
ordinal_features = ['Profession','Education']
# Fit and transform the specific ordinal features
data[ordinal_features] = encoder.fit_transform(data[ordinal_features])
```

**Apply K-Means algorithm**

```python
from sklearn.cluster import KMeans
```

```python
Kmeans_features=['House Loan', 'Wife Working','Personal loan','Age', 'No of Dependents', 'Total Salary', 'Minimum Price to buy a EV_Car']
data1=data[Kmeans_features]
```

```python
# applying the kmeans clustering
wcss=[]
for k in range(1,11):
    #preventing random initialization: 'init=k-means++'
    kmeans= KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(data1)
    wcss.append(kmeans.inertia_)
```

```python
# findout the optimal number of clusters by using elbow method
plt.figure(figsize=(10,6))
plt.plot(range(1,11),wcss, linestyle='solid', marker='o',
          markersize=5)
plt.title('Elbow Method', size=20)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# after finding the optimal clusters apply
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
kmeans.fit(data1)
labels=kmeans.labels_
data1['cluster'] = labels
print(kmeans.cluster_centers_)
```

```python
pca2=PCA(n_components=3)
pca_com=pca2.fit_transform(data1)
```

```python
names=['pc1','pc2','pc3']
frame=pd.DataFrame(data=pca_com,columns=names)
```

```python
# Visulazing clusters
sns.scatterplot(data=frame, x="pc1", y="pc2", hue=kmeans.labels_,s = 100)
plt.legend()
plt.show()
```

Getting maximum and minimum values in every cluster

**Calculating the sillihoute score**

```python
from sklearn.metrics import silhouette_score
# Calculate and print the Silhouette Score
silhouette_avg = silhouette_score(data1,labels)
print(f'Silhouette Score: {silhouette_avg}')
```

From the score,clustering of the data were almost divide accurately

```python
grouped = data1.groupby('cluster')['Minimum Price to buy a EV_Car']
max_values = grouped.max()
min_values = grouped.min()
```

```python
print(max_values),print(min_values)
```
